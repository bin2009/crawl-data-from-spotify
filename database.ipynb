```python
import os
import pandas as pd
from unidecode import unidecode

# Đường dẫn đến thư mục chứa các file CSV
merge_dir = 'D:/DUT/NAM4-KI1/PBL6/CRAWL/crawl-data-from-spotify/Merge'
array_artist = set()  # Sử dụng set để tránh trùng lặp

# Liệt kê tất cả các file trong thư mục
files = [f for f in os.listdir(merge_dir) if f.endswith('.csv')]

# Đọc và xử lý từng file CSV
for file in files:
    print(f"Đang xử lý file: {file}")
    file_path = os.path.join(merge_dir, file)
    df = pd.read_csv(file_path)
    
    # Lặp qua từng hàng trong DataFrame
    for index, row in df.iterrows():
        if '||' in row['Artists']:
            artists = row['Artists'].split('||')
            for art in artists:
                art = art.split('(')[0].strip()
                art = unidecode(art)  # Chuyển sang dạng chữ không dấu
                array_artist.add(art)  # Thêm nghệ sĩ vào set để tránh trùng lặp
        else:
            artists = row['Artists'].split('(')[0].strip()
            artists = unidecode(artists)  # Chuyển sang dạng chữ không dấu
            array_artist.add(artists)

# Chuyển đổi set thành danh sách
array_artist = list(array_artist)

# Tạo DataFrame từ danh sách nghệ sĩ
df_artists = pd.DataFrame(array_artist, columns=['Name'])

# Đường dẫn đến file Excel để xuất
output_excel_path = 'D:/DUT/NAM4-KI1/PBL6/CRAWL/crawl-data-from-spotify/Data-Artist/AAA.xlsx'

# Xuất DataFrame ra file Excel
df_artists.to_excel(output_excel_path, index=False)

print(f"Danh sách các nghệ sĩ đã được xuất ra file: {output_excel_path}")
```

```python
import pandas as pd
import uuid

# Load artist information
nameArtist = 'SonTungMTP'
artists = pd.read_csv('D:/DUT/NAM4-KI1/PBL6/CRAWL/crawl-data-from-spotify/Data-Artist/inf.csv')

# SQL INSERT template for the Artist table
insert_sql_artist = 'INSERT INTO "Artist" ("id", "name", "avatar", "bio", "createdAt", "updatedAt") VALUES\n'

# Prepare artist insert values and a mapping between id2 and id (UUID)
values_artist = []
artist_id_map = {}
for index3, row3 in artists.iterrows():
    artist_uuid = str(uuid.uuid4()) 
    artist_id2 = row3['Artist ID']
    name = row3[' Artist Name'].strip()
    avatar = row3[' Images'].split('||')[1].split('(')[0].strip()
    bio = 'Null'
    values_artist.append(f"('{artist_uuid}', '{name}', '{avatar}', '{bio}', NOW(), NOW())")
    
    # Store the mapping for use later
    artist_id_map[artist_id2] = artist_uuid




# Load album and track data for the artist
df = pd.read_csv(f'D:/DUT/NAM4-KI1/PBL6/CRAWL/crawl-data-from-spotify/Merge/{nameArtist}.csv')

# SQL INSERT templates for Album, AlbumImage, Song, and ArtistSong tables
insert_sql_album = 'INSERT INTO "Album" ("albumId",  "title", "releaseDate",  "coverImage", "albumType", "createdAt", "updatedAt") VALUES\n'
insert_sql_album_image = 'INSERT INTO "AlbumImage" ("albumId", "image", "size", "createdAt", "updatedAt") VALUES\n'
insert_sql_track = 'INSERT INTO "Song" ("id", "albumId", "title", "duration", "lyric", "filePathAudio", "privacy", "uploadUserId", "createdAt", "updatedAt") VALUES\n'
insert_sql_artist_song = 'INSERT INTO "ArtistSong" ("songId", "artistId", "main", "createdAt", "updatedAt") VALUES\n'

# Prepare insert values for albums, album images, tracks, and artist-song relationships
values_album = []
values_album_image = []
values_track = []
values_track_artist = []

# Remove duplicate albums
unique_album = df.drop_duplicates(subset=['AlbumID'])

for index, row in unique_album.iterrows():
    # Album details
    albumId = str(uuid.uuid4()) 
    albumIdTemp = row['AlbumID']
    
    values_album.append(f"('{albumId}', '{row['Album']}', '{row['ReleaseDate']}', '{row['AlbumImage']}','{row['AlbumType'].lower()}', NOW(), NOW())")

    # Album images
    images = row['AlbumImage'].split('||')
    for img in images:
        img_url, img_size = img.split('(')
        img_url = img_url.strip()
        img_size = img_size.strip().split('size:')[1].strip().split(')')[0].strip()
        values_album_image.append(f"('{albumId}', '{img_url}', '{img_size}', NOW(), NOW())")

    # Tracks associated with the current album
    tracks_in_album = df[df['AlbumID'] == albumIdTemp]
    for _, track in tracks_in_album.iterrows():
        trackId = str(uuid.uuid4())
        values_track.append(f"('{trackId}', '{albumId}', '{track['TrackName']}', '{track['TrackDuration']}', 'lyric', '{track['audio_path']}', FALSE, 'Null', NOW(), NOW())")

        # Associate track with artists
        artistNames = track['ArtistsName'].split('||')
        
        # Handle artist-song relation based on artistNames length
        if len(artistNames) == 1:
            # Only one artist, set main to TRUE
            artist_name = artistNames[0]
            artId2 = artist_name.split('(ID:')[1].split(')')[0].strip()
            artistId = artist_id_map.get(artId2, 'Unknown')  # Use artist_id_map to fetch the UUID
            values_track_artist.append(f"('{trackId}',  '{artistId}', , TRUE, NOW(), NOW())")
        else:
            # Multiple artists, first one is main, others are not
            for i, artist in enumerate(artistNames):
                artId2 = artist.split('(ID:')[1].split(')')[0].strip()
                artistId = artist_id_map.get(artId2, 'Unknown')  # Fetch UUID from the map
                main = 'TRUE' if i == 0 else 'FALSE'
                values_track_artist.append(f"('{trackId}', '{artistId}', {main}, NOW(), NOW())")

# Combine values for SQL insert queries
insert_sql_artist += ",\n".join(values_artist) + ";"
insert_sql_album += ",\n".join(values_album) + ";"
insert_sql_album_image += ",\n".join(values_album_image) + ";"
insert_sql_track += ",\n".join(values_track) + ";"
insert_sql_artist_song += ",\n".join(values_track_artist) + ";"

# Save the SQL queries to a file
sql_file_path = f'D:/DUT/NAM4-KI1/PBL6/CRAWL/crawl-data-from-spotify/SQL2/{nameArtist}_sql.sql'

with open(sql_file_path, 'w', encoding='utf-8') as file:
    file.write(insert_sql_artist)
    file.write("\n\n\n\n")
    file.write(insert_sql_album)
    file.write("\n\n\n\n")
    file.write(insert_sql_album_image)
    file.write("\n\n\n\n")
    file.write(insert_sql_track)
    file.write("\n\n\n\n")
    file.write(insert_sql_artist_song)
```
